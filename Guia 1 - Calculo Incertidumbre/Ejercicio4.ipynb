```python
# INCERTIDUMBRE TIPO B (CORRIENTE)

e_lectura_I = 0.1           #error de lectura

error_total_I = (10 * e_lectura_I)/100

uj_I = error_total_I/2

#se divide por 2 por el 95% de confianza
#(ver tabla de distribucin normal --> 2 sigmas)

print("Incertidumbre tipo B de la corriente:    uj_I = %0.4f" %uj_I)
```

```python
# INCERTIDUMBRE TIPO B (VOLTIMETRO)

#Como el multmetro es de 4 digitos y 3/4, tenemos 4 digitos que pueden variar de 0 a 9, 
#y el dgito ms significativo solo puede variar entre 0 y 3. 
#Es decir que el rango de valores posibles es de 00000 a 39999 (ubicando la coma donde corresponda).

#Uso escala de 400V --> que vare un digito me varia 0.01V

medicion_V = 123.38
error_lectura_V = 0.04               
digitos_V = 1
error_absoluto_V = digitos_V * 0.01

error_total_V = (error_lectura_V/100)* medicion_V + error_absoluto_V

uj_V = error_total_V/np.sqrt(3)             #calculo incertidumbre tipo B

#Supongo Distribucin estandar (rectangular) --> divido 
#el error por raiz cuadrada de 3

print("Error total: " ,error_total_V)
print("Incertidumbre tipo B:    uj_V = %0.10f" %uj_V)
```
