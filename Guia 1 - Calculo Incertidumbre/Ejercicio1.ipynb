Como el multímetro es de 3 digitos y medio, tenemos 3 digitos que pueden variar de 0 a 9, y el dígito más significativo solo puede variar entre 0 y 1. Es decir que el rango de valores posibles es de 0000 a 1999 (ubicando la coma donde corresponda).

En el rango de 200V, el tester muestra a fondo de escala 199.9V --> Con 3 digitos de error, el tester puede mostrar 199.6V en vez de 199.9V  --> El error absoluto es de 0.3V

```python
import numpy as np
```

```python
# EVALUACION TIPO A

V_med = 100.145              #valor medio
V_std = 1.489                #desvio estandar

n = 20

ui = V_std
#ui = V_std/np.sqrt(n)        #calculo incertidumbre tipo A

print("Incertidumbre tipo A:     ui = %0.10f" %ui)
```

```python
# EVALUACION TIPO B

error_lectura = 0.5
medicion = 100.1                        # No usamos 100.145 ya que no dan los digitos
digitos = 3
error_absoluto = digitos * 0.1

error_total = (error_lectura/100)* medicion + error_absoluto

#Supongo Distribución estandar (rectangular) --> divido 
#el error por raiz cuadrada de 3 

uj = error_total/np.sqrt(3)             #calculo incertidumbre tipo B

print("Error total: " ,error_total)
print("Incertidumbre tipo B:    uj = %0.10f" %uj)
```

```python
# INCERTIDUMBRE COMBINADA

uc = np.sqrt(ui**2 + uj**2)
print("Incertidumbre combinada:     uc = %0.10f" %uc)
```

```python
# GRADOS EFECTIVOS DE LIBERTAD

v_eff = (uc**4) / (ui**4/(n-1))           #calculo grados efectivos de libertad 
print("Grados de libertad: %d" %v_eff)    #lo casteamos a entero
```

```python
# INCERTIDUMBRE EXAPNDIDA

#Con menos de 30 grados de libertad aplica la t-student, ya que no tengo 
#suficientes grados de libertad para llegar a la normal (gausseana).

#El coeficiente K lo obtengo de la tabla de student (ver en PPT).
#En la tabla vas a la columna de  95 y la entrada de 7 grados de libertad,
#obteniendo K = 2.09

k = 2.09

U = k * uc    #incertidumbre expandida

print("Incertidumbre combinada expandida:     U = %0.10f" %U)

print("El resultado de la medicion es 100.145V +/- %0.5fV" %U ,"con confianza de 95%")
```
