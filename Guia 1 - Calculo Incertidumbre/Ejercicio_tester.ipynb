# Ejercicio medici√≥n de tester

Error = 2% of read + 10 digitos

```python
import numpy as np
```

```python
i = [10.01, 10.00, 10.02, 10.01, 10.00]
i = np.array(i)

i_mean = i.mean()              #calculo valor medio
i_std = i.std(ddof = 1)        #calculo desvio estandar
n = len(i)

print(i_mean, i_std, n)

i_x = i_mean
```

```python
# EVALUACION TIPO A
ui_x = i_std/np.sqrt(n)   

# EVALUACION TIPO B                                  
uj_x = ((2/100)*i_x + (10/1000)*10)/np.sqrt(3)          

# INCERTIDUMBRES COMBINADAS
uc_x = np.sqrt(ui_x**2 + uj_x**2)

# GRADOS EFECTIVOS DE LIBERTAD
v_eff = (uc_x**4) / (ui_x**4/(n-1)) 

# Ratio tipoA/tipoB
r = ui_x/uj_x

print("Incertidumbre Tipo A:", ui_x)
print("Incertidumbre Tipo B:", uj_x)
print("Incertidumbre combinada:", uc_x)
print("Grados de libertad: %d" %v_eff)    #lo casteamos a entero
print("Ratio tipoA/tipoB:", r)
```

```python
# INCERTIDUMBRE EXPANDIDA
k = 1.65

U = k * uc_x      #incertidumbre expandida

print("Incertidumbre total: %.4f" %uc_x)
print("Incertidumbre expandida: %.4f" %U)
```
