```python
import numpy as np
```

## Incertidumbre tipo A Voltímetro

Al ser una única medición, no se puede estimar una incertidumbre tipo A. 

## Incertidumbre tipo B Generador

```python
# ESTUDIO TIPO B

# Asumimos distribución uniforme --> dividimos por sqrt(3)

Vref_uj = (0.02 * 1)/100 / np.sqrt(3)

print("Incertidumbre tipo B del generador:  %0.8f"  %Vref_uj)
```

## Error de método de la medición

```python
Vi = 1.007

Rv = 10e6
Ri = 0.5

Em = Rv/(Rv+Ri) - 1

V_ok = Vi/(1+Em)
Delta_Em = Em * V_ok

print("Error de método:  %0.10f" %Em)
print("Error absoluto de método:  %0.10f" %Delta_Em)
```

## Comparación Error de método vs incertidumbre

Vemos que la incertidumbre tipo B tiene 4 órdenes de magnitud más que el error de método, por lo que no hay que tener en cuenta este último. Entonces no corregimos.

## Error de la medición

```python
Vref = 1
C = Vi - Vref

dC_dVref = -1

C_uc = np.sqrt(dC_dVref**2 * Vref_uj**2) #La incertidumbre termina siendo la misma que la de el generador de referencia (Distribución uniforme)
```

## Resultado, con incertidumbre expandida

Para exandir la incertidumbre de distribución uniforme, vuelvo a multiplicar por $\sqrt{3}$ para pasarlo al semirrango, y luego por 0.95 para obtener el intervalo que asegura un 95% de confianza.

```python
C_uc_exp = C_uc * np.sqrt(3) * 0.95

print("El error en la tensión medida es de %0.3fV +/- %0.4fV @ 95.45%% confianza" %(C,C_uc_exp))
```
