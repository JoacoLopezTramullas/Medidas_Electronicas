#### Calculo error de metodo y su incertidumbre

Tomando el error de metodo como funcion de medicion el error de metodo es:

```python
import numpy as np

f = 15e3
w = 2*np.pi*f
Rg=600
Cd = 150e-12
Cv = 220e-12
Rv = 1e6

em = np.sqrt((w**2+((Rg/Rv)/(Rg*(Cd+Cv)))**2)/(w**2+((Rg/Rv+1)/(Rg*(Cd+Cv)))**2))
```

Como la frecuencia del generador tiene una dispersion, el error de mtodo tambien tendr su incertidumbre. Donde el delta de frecuencia ser y la incertidumbre TIPO B u_j_f = u_c_f (asumiendo distribucion uniforme):

```python
delta_f = 100/1e6 * f

u_c_f = delta_f / np.sqrt(3)
```

```python
## incertidumbre combinada del error de metodo

dem_df = (w**2*1/(Rg*(Cv+Cd)))/(np.sqrt(w**2+(Rg/Rv)/(Rg*(Cd+Cv)))*(w**2+(Rg/Rv+1)/(Rg*(Cd+Cv)))**1.5)

u_c_em = dem_df * u_c_f

print('El error de metodo es =  %f\nLa incertidumbre del error de metodo es = %f' %(em,u_c_em))
```

#### Incertidumbre Voltimetro

Como el voltimetro midio 1.8562V se asume que se midio en la escala de 2v. Por lo tanto el error debido a los 2 digitos ser de 0.0002V. El error total TIPO B del instrumento (asumiendo distribucion uniforme) ser:

```python
Vi = 1.8562
delta_v = 0.05/100*Vi+0.0002
u_c_v = delta_v/np.sqrt(3)

print('La incertidumbre combinada sin expandir del voltimetro es = %fV' %(delta_v))
```

Notar que la distribucion del voltimetro es la misma que la del estudio tipo B, es decir, Uniforme

### Comparacion Error de Metodo y Error de Lectura

```python
Err_V = delta_v/Vi

print('Error relativo de la medicion %f' %Err_V)
```

Como el error de metodo tiene su incertidumbre, pero el valor promedio de Em ya es 10 veces mayor al error del voltimetro, se corrije. Adems como la correcin del error de mtodo es simplemente multiplicar por una constante, la distribucin de la incertidumbre combinada de los valores corregidos sigue siendo uniforme, por lo que no hace falta recalcular los grados de libertad para saber que tipo de distribucion posee.

```python
V_ok = Vi/(em+1)

dV_ok_Vi = 1/(em+1)

V_ok_uc = dV_ok_Vi * u_c_v
```

Siendo la distribucion de la tension medida uniforme, para una cobertura del 95%, el factor de expansion debe ser $K = \sqrt{3} * 0.95 = 1.65$ 

```python
V_ok_uc_exp = V_ok_uc * np.sqrt(3) * 0.95
```

```python


print("'El valor obtenido de la medicin es = %0.3f +- %0.3f V @ 95%% confianza (K = 1.65)" %(V_ok, V_ok_uc))
```
