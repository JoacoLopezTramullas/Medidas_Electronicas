# Ejercicio 4





Las corrientes medidas son del orden de los ampres, por lo que la escala de corriente utilizada es de 10A. La escala mnima de tensin es de 200mV, por lo que la resistencia del ampermetro estimada es de $\frac{200mV}{10A} = 20m\Omega$

Para saber si es necesario corregir el error de mtodo, primero calculamos la incertidumbre de la resistencia, la expandimos a un 99% de confianza y la comparamos con el error de mtodo multiplicado por R real (para que pasarlo a error absoluto y no porcentual).

```python
import numpy as np

Vi = np.array([10.11, 10.21, 10.10, 10, 10.19])
Ii = np.array([2.51, 2.52, 2.52, 2.49, 2.49])
```

## Ampermetro

```python
#INCERTIDUMBRE TIPO A
Ii_mean = Ii.mean()              #calculo valor medio
Ii_std = Ii.std(ddof = 1)        #calculo desvio estandar
Ii_n = len(Ii)                   #calculo cantidad de muestras

Ii_ui = Ii_std/np.sqrt(Ii_n)     #calculo incertidumbre tipo A


#NCERTIDUMBRE TIPO B
#Rango 10A: resolucin: 10mA, precisin: +/-2.5% de lectura + 5 dgitos
error_amp = Ii_mean * 2.5/100 + 5*10e-3

#Asumiendo distribucin uniforme:
Ii_uj = error_amp / np.sqrt(3)


#NCERTIDUMBRE COMBINADA
Ii_uc = np.sqrt(Ii_ui**2 + Ii_uj**2)

print("Incertidumbre tipo A de ampermetro:     Ii_ui = %0.8f" %Ii_ui)
print("Incertidumbre tipo B de ampermetro:     Ii_uj = %0.8f" %Ii_uj)
print("Incertidumbre combinada de ampermetro:  Ii_uc = %0.8f" %Ii_uc)
```

## Voltmetro

```python
#INCERTIDUMBRE  TIPO A
Vi_mean = Vi.mean()              #calculo valor medio
Vi_std = Vi.std(ddof = 1)        #calculo desvio estandar
Vi_n = len(Vi)                   #calculo cantidad de muestras

Vi_ui = Vi_std/np.sqrt(Vi_n)     #calculo incertidumbre tipo A


#INCERTIDUMBRE TIPO B
#Rango 20V: resolucin: 10mV, precisin: +/-0.5% de lectura + 2 dgitos
error_volt = Vi_mean * 0.5/100 + 2*10e-3

#Asumiendo distribucin uniforme:
Vi_uj = error_volt / np.sqrt(3)


#INCERTIDUMBRE COMBINADA
Vi_uc = np.sqrt(Vi_ui**2 + Vi_uj**2)

print("Incertidumbre tipo A de voltmetro:     Vi_ui = %0.8f" %Vi_ui)
print("Incertidumbre tipo B de voltmetro:     Vi_uj = %0.8f" %Vi_uj)
print("Incertidumbre combinada de voltmetro:  Vi_uc = %0.8f" %Vi_uc)
```

## Resistencia Indicada

```python
#INCERTIDUMBRE DE RESISTENCIA INDICADA

#Clculo de Ri y derivadas
Ri_mean = Vi_mean / Ii_mean

dR_dV = 1/Ii_mean
dR_dI = -Vi_mean / Ii_mean**2

#Clculo coef de correlacin
Med_mtx = np.array([Vi,Ii])

r = np.corrcoef(Med_mtx)
r_V_I = r[1,0]

#INCERTIDUMBRE COMBINADA
Ri_uc = np.sqrt((dR_dV * Vi_uc)**2 + (dR_dI * Ii_uc)**2 + 2 * dR_dV * dR_dI * r_V_I * Ii_uc *  Vi_uc)

print("Resistencia medida:        Ri = %0.5f" %Ri_mean)
print("Incertidumbre combinada:   Ri_uc = %0.5f" %Ri_uc)
```

## Error de Mtodo - Corriente Bien medida

```python
#ERROR DE MTODO - CORRIENTE BIEN MEDIDA
RA = 200e-3/10 #Estimada a partir de la mnima escala de tenson y la escala de corriente empleada
Em = RA / (Ri_mean-RA)

Delta_Em = Em * (Ri_mean-RA) #En este caso el error absoluto es RA

print("Error absoluto de mtodo: %0.3f" %Delta_Em)
```

## Comparacin de incertidumbre de la Ri con el error de mtodo

Antes de comparar nos parece razonable asegurar un 99% de confianza en la incertidumbre del a Ri.

Como las incertidumbres combinadas del voltmetro y del ampermetro son comparables, suponemos que al convolucionar 4 incertidumbres la distribucin de la incertidumbre de la resistencia es gaussiana. (TEOREMA CENTRAL DEL LMITE).

Por esta razn, expandimos la incertidumbre con un coeficiente de 3.

```python
# Incertidumbre expandida de Ri
Ri_uc_exp = Ri_uc * 3

print("Incertidumbre expandida: %0.5f" %Ri_uc_exp)
```

Como se puede ver, la incertumbre expandida (error de lectura) es aproximadamente 15 veces mayor al error de mtodo. Por esta razn, NO se lo debe corregir.

```python
print("Resistencia de canal RDSon = %0.1f +- %0.1f @ 99%% confianza (K = 3)" %(Ri_mean, Ri_uc_exp))
```
