## Mediciones

```python
import numpy as np

#Defino vectores
Vi = np.array([22.323, 22.325, 22.320, 22.331, 22.332, 22.330, 22.327, 22.323, 22.329, 22.325])
Ii = np.array([145.1, 145.2, 144.9, 145.7, 145.2, 146.3, 145.3, 145.1, 145.9, 145.4])*1e-3

#Calculo cantidad de muestras
n = len(Vi)

#Calculo valores medios
Vi_mean = Vi.mean()   
Ii_mean = Ii.mean()

#Calculo desvíos estándares 
Vi_std = Vi.std(ddof = 1)       
Ii_std = Ii.std(ddof = 1)        

print("Valor medio de la tension:    %0.4fV" %Vi_mean)
print("Valor medio de la corriente:  %0.5fA" %Ii_mean)
```

## Voltímetro

```python
#INCERTIDUMBRES VOLTÍMETRO

#------- TIPO A -------
Vi_ui = Vi_std/np.sqrt(n)


#------- TIPO B -------
# +/- (0,05 % lectura + 2 dígitos)        Rv = 10M
resolucion = 1e-3
error_volt = Vi_mean * 0.05/100 + 2*resolucion

#Asumiendo distribución uniforme (dividimos por sqrt(3)):
Vi_uj = error_volt / np.sqrt(3)


#------- COMBINADA -------
Vi_uc = np.sqrt(Vi_ui**2 + Vi_uj**2)


#------- GRADOS DE LIBERTAD -------
Vi_eff = (Vi_uc**4) / (Vi_ui**4/(n-1)) 

print("Incertidumbre tipo A de voltímetro:     Vi_ui = %0.8f" %Vi_ui)
print("Incertidumbre tipo B de voltímetro:     Vi_uj = %0.8f" %Vi_uj)
print("Incertidumbre combinada de voltímetro:  Vi_uc = %0.8f" %Vi_uc)
print("Grados de libertad de Ri:  %d" % Vi_eff)
```

## Amperímetro

```python
#INCERTIDUMBRES AMPERÍMETRO

#------- TIPO A -------
Ii_ui = Ii_std/np.sqrt(n)


#------- TIPO B -------

#Calculamos error cometido por el instrumento
#Clase 0.2 --> 0.2 = (error absoluto max / fondo de escala)*100
fondo_escala = 150e-3;
error_amp = (0.2/100) * fondo_escala 

#Calculamos error de lectura (instrumento analógico)
#Suponemos un error de lectura de 0.2div
#Como tenemos 150 divisiones, y lo máximo que se puede medir es 150 mA --> 1div = 1mA
#Como la corriente media da 145.41mA --> Estoy usando 145.41 divisiones
resolucion = 1e-3
divisiones = Ii_mean/resolucion
error_lectura = (0.2/divisiones)     # esta en porcentaje

error_total = error_lectura * Ii_mean + error_amp

#Asumiendo distribución uniforme (dividimos por sqrt(3)):
Ii_uj = error_total / np.sqrt(3)


#------- COMBINADA -------
Ii_uc = np.sqrt(Ii_ui**2 + Ii_uj**2)


#------- GRADOS DE LIBERTAD -------
Ii_eff = (Ii_uc**4) / (Ii_ui**4/(n-1)) 

print("Incertidumbre tipo A de amperímetro:     Ii_ui = %0.8f" %Ii_ui)
print("Incertidumbre tipo B de amperímetro:     Ii_uj = %0.8f" %Ii_uj)
print("Incertidumbre combinada de amperímetro:  Ii_uc = %0.8f" %Ii_uc)
print("Grados de libertad de Ri:  %d" % Ii_eff)
```

Vemos que los grados efectivos de libertad tanto en el amperímetro como en el voltímetro son mayores a 30, por lo que en ambos casos domina la distribución Tipo B. 

## Resistencia indicada

```python
#INCERTIDUMBRE DE RESISTENCIA INDICADA

#Calculo de Ri
Ri_mean = Vi_mean / Ii_mean

#Calculo derivadas
dR_dV = 1/Ii_mean
dR_dI = -Vi_mean / Ii_mean**2

#Calculo coef de correlación
Med_mtx = np.array([Vi,Ii])
r = np.corrcoef(Med_mtx)
r_V_I = r[1,0]                  

#INCERTIDUMBRE COMBINADA
Ri_uc = np.sqrt((dR_dV * Vi_uc)**2 + (dR_dI * Ii_uc)**2 + 2 * dR_dV * dR_dI * r_V_I * Ii_uc *  Vi_uc)

#GRADOS DE LIBERTAD DE RI
veff_Ri = Ri_uc**4 * (n-1) /((dR_dV * Vi_ui)**4+(dR_dI * Ii_ui)**4)

print("Resistencia medida:        Ri = %0.8f" %Ri_mean)
print("Incertidumbre combinada:   Ri_uc = %0.8f" %Ri_uc)
print("Grados de libertad de Ri:  %d" % veff_Ri)
```

Al ser la distribución del voltímetro y la del amperímetro tipo B, la distribución de la resistencia medida (Ri) va a ser rectangular, ya que estamos convolucionando 2 rectangulares. Es por esto que usamos la siguiente tabla para determinar el coeficiente de expansión con un factor de cobertura del 95%:



```python
Ratio = Ii_uc / Vi_uc

print("Ratio:  %0.8f" %Ratio)
```

Viendo la tabla, usamos un coeficiente de expansión K = 1.65

```python
#Incertidumbre expandida
Ri_uc_exp = Ri_uc * 1.65

Ri_uc_exp_porc = (Ri_uc_exp/Ri_mean)*100

print("Incertidumbre expandida de Ri (valor absoluto): %0.5f" %Ri_uc_exp)
print("Incertidumbre expandida de Ri (porcentual):     %0.5f%%" %Ri_uc_exp_porc)
```

## Error de método - Tensión bien medida

```python
#ERROR DE MÉTODO - TENSIÓN BIEN MEDIDA
Rv = 10e6 
Em = -Ri_mean/Rv

Em_porc = Em*100 

print("Error de método (porcentual): %0.8f%%" %Em_porc)
```

## Comparación de incertidumbre de la Ri con el error de método


Como se puede ver, la incertumbre expandida (error de lectura) es aproximadamente 212 veces mayor al error de método. Por esta razón, no se lo debe tener en cuenta.

## Conclusión

```python
print("Resistencia medida = %0.1f +- %0.1f @ 99%% confianza (K = 3)" %(Ri_mean, Ri_uc_exp))
```
